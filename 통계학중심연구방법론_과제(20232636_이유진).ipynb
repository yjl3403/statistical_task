주제: 성별에 따라 음악을 들으면서 공부했을 때 점수 차이가 있을까?

```python
from google.colab import files
import pandas as pd
from scipy.stats import shapiro, ttest_ind
import io

# 1. 파일 업로드
uploaded = files.upload()

# 2. CSV 파일 불러오기
df = pd.read_csv(io.BytesIO(uploaded['통계학과제.csv']))

# 3. 성별에 따른 점수 분리
male_scores = df[df["성별"] == "남"]["점수"]
female_scores = df[df["성별"] == "여"]["점수"]

# 4. Shapiro-Wilk 정규성 검정
shapiro_male = shapiro(male_scores)
shapiro_female = shapiro(female_scores)

print("Shapiro-Wilk Test 결과:")
print(f"남학생 정규성 검정: W = {round(shapiro_male.statistic, 4)}, p-value = {round(shapiro_male.pvalue, 4)}")
print(f"여학생 정규성 검정: W = {round(shapiro_female.statistic, 4)}, p-value = {round(shapiro_female.pvalue, 4)}")

# 5. Unpaired Student’s t-test
ttest_result = ttest_ind(male_scores, female_scores, equal_var=False)

print("Unpaired Student's t-test 결과:")
print(f"t-statistic = {round(ttest_result.statistic, 4)}, p-value = {round(ttest_result.pvalue, 4)}")

# 6. 결론
alpha = 0.05

if ttest_result.pvalue > alpha:
    print("결론: 귀무가설 채택 -> 성별에 따른 점수 차이가 통계적으로 유의하지 않습니다.")
else:
    print("결론: 귀무가설 기각 -> 성별에 따른 점수 차이가 통계적으로 유의미합니다.")

```

##### 귀무가설(H0): 음악을 들으면서 공부했을 때 남학생과 여학생의 평균 점수에는 차이가 없다.
##### 대립가설(H1): 음악을 들으면서 공부했을 때 남학생과 여학생의 평균 점수에는 차이가 있다.
##### 해석: 정규성 검정 결과, 남학생과 여학생 모두 정규성을 만족했다. 따라서 Unpaired Student's t-test를 실시한 결과, p-value가 유의수준 0.05보다 크므로 귀무가설을 기각할 수 없다. 결과적으로 성별에 따른 음악을 들으면서 공부했을 때의 점수 차이는 통계적으로 유의하지 않다.
